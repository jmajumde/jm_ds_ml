```python
import warnings
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

warnings.filterwarnings('ignore')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.layers.convolutional import Conv1D, MaxPooling1D
```

```python
series = pd.read_csv('monthly-car-sales.csv', header=0, index_col=0)
series.shape
```

A line plot is created. The dataset has an obvious trend and seasonal component. The period of the seasonal component could be six months or 12 months.

```python
# split the univariate data into train and test set 
n_test=12 # keeping 12 months sales data aside for test

def train_test_split(data, n_test):
    return data[:-n_test], data[-n_test:]
```

```python
# make the series into supervised learning format 
# for 3 years 3*12 = 36 months data

n_in=36
n_out=1
df = pd.DataFrame(series)
print(df.head())
train, test = train_test_split(df, n_test=n_test)
print(train.shape, test.shape)

cols=list()

# seq from t-n... t-1
for i in range(n_in, 0, -1):
    cols.append(df.shift(i))
    
for i in range(0, n_out):
    cols.append(df.shift(-i))

# concat cols
agg = pd.concat(cols, axis=1)
agg

# drop null
agg.dropna(inplace=True)
agg
```

```python

# putting this together and create a function for series_to_supervised

def series_to_supervised(data, n_in=1, n_out=1):
    cols=list()
    df = pd.DataFrame(data)
    # input seq from t-n... t-1
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
    
    # forecast seq t, t+1, .. t+n
    for i in range(0, n_out):
        cols.append(df.shift(-i))

    # concat cols
    agg = pd.concat(cols, axis=1)
    
    # drop null
    agg.dropna(inplace=True)
    
    return agg.values
    
```

```python
# create MLP model
def mlp_model_fit(train, config):
    # unpack config object
    n_input, n_nodes, n_epochs, n_batch = config
    
    # prepare data
    data = series_to_supervised(train, n_input)
    
    # train_x, train_y
    train_x, train_y = data[:,:-1], data[:,-1]
    #print("train_x len:", len(train_x), "train_y len:", len(train_y))
    
    # define model
    model = Sequential()
    model.add(Dense(n_nodes, input_dim=n_input, activation='relu'))
    model.add(Dense(1))
    model.compile(loss='mse', optimizer='adam')
    
    #print(model.summary())
    
    # fit the model
    #model.fit(train_x, train_y, epochs=n_epochs, batch_size=n_batch, verbose=2)
    model.fit(train_x, train_y, epochs=n_epochs, batch_size=n_batch, verbose=0)
    
    return model
    
```

CNN can support multiple features or types of observations at each time step, which are interpreted as channels of an image. We only have a single feature at each time step, therefore the required three-dimensional shape of the input data will be [n_samples, n_input, 1].

```python
# create model with 1D Convolutional Network


def conv1d_model_fit(train, config):
    # unpack config object
    n_input, n_filters, n_kernel, n_epochs, n_batch = config
    
    # prepare data
    data = series_to_supervised(train, n_input)
    
    # train_x, train_y
    train_x, train_y = data[:,:-1], data[:,-1]
    train_x = train_x.reshape((train_x.shape[0], train_x.shape[1], 1))
    
    # define model
    model = Sequential()
    model.add(Conv1D(filters=n_filters, kernel_size=n_kernel, activation='relu', input_shape=(n_input, 1)))
    model.add(Conv1D(filters=n_filters, kernel_size=n_kernel, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))
    model.add(Flatten())
    model.add(Dense(1))
    model.compile(loss='mse', optimizer='adam')
    
    #print(model.summary())
    
    # fit the model
    model.fit(train_x, train_y, epochs=n_epochs, batch_size=n_batch, verbose=2)
    
    return model
```

```python
def conv1d_model_predict(model, history, config):
    # unpack config
    n_input, _, _, _, _ = config
    # prepare data
    x_input = np.array(history[-n_input:]).reshape((1, n_input, 1))
    # forecast
    yhat = model.predict(x_input, verbose=0)
    return yhat[0]


```

```python
def mlp_model_predict(model, history, config):
    # unpack config
    n_input, _, _, _ = config
    
    # prepare data
    x_input = np.array(history[-n_input:]).reshape(1, n_input)
    # forecast
    yhat = model.predict(x_input, verbose=0)
    return yhat[0]
```

The `predict()` function expects one or more samples of inputs when making a prediction, so providing a single sample requires the array to have the shape `[1, n_input]` for MLP model predict function above, where `n_input` is the number of time steps that the model expects as input.

Similarly, the predict() function returns an array of predictions, one for each sample provided as input. In the case of one prediction, there will be an array with one value.

The model_predict() function above implements this behavior, taking the model, the prior observations, and model configuration as arguments, formulating an input sample and making a one-step prediction that is then returned.

##### Repeat Evaluation
Neural network models are `stochastic`.

This means that, given the same model configuration and the same training dataset, a different internal set of weights will result each time the model is trained that will in turn have a different performance.

This is a benefit, allowing the model to be adaptive and find high performing configurations to complex problems.

It is also a problem when evaluating the performance of a model and in choosing a final model to use to make predictions.

To address model evaluation, we will evaluate a model configuration multiple times via `walk-forward validation` and report the error as the average error across each evaluation.

This is not always possible for large neural networks and may only make sense for small networks that can be fit in minutes or hours.

The `repeat_evaluate()` function below implements this and allows the number of repeats to be specified as an optional parameter that defaults to 30 and returns a list of model performance scores: in this case, RMSE values.

```python

# summarize model performance
def summarize_scores(name, scores):
	# print a summary
	scores_m, score_std = np.mean(scores), np.std(scores)
	print('%s: %.3f RMSE (+/- %.3f)' % (name, scores_m, score_std))
	# box and whisker plot
	plt.boxplot(scores)
	plt.show()
```

```python
def measure_rmse(actual,predicted):
    return np.sqrt(mean_squared_error(actual, predicted))
```

```python
def mlp_walk_forward_validation(data, n_test, cfg):
    predictions = list()
    # split dataset
    train, test = train_test_split(data, n_test)
    #print(train.shape, test.shape)
    
    # fit model
    model = mlp_model_fit(train, cfg)
    # seed history with training dataset
    history = [x for x in train]
    #print("history len:", len(history))
    
    # step over each time-step in the test set
    for i in range(len(test)):
        # fit model and make forecast for history
        yhat = mlp_model_predict(model, history, cfg)
        # store forecast in list of predictions
        predictions.append(yhat)
        # add actual observation to history for the next loop
        history.append(test[i])
    # estimate prediction error
    error = measure_rmse(test, predictions)
    print(' > %.3f' % error)
    return error


```

```python
def conv1d_walk_forward_validation(data, n_test, cfg):
	predictions = list()
	# split dataset
	train, test = train_test_split(data, n_test)
	# fit model
	model = conv1d_model_fit(train, cfg)
	# seed history with training dataset
	history = [x for x in train]
	# step over each time-step in the test set
	for i in range(len(test)):
		# fit model and make forecast for history
		yhat = conv1d_model_predict(model, history, cfg)
		# store forecast in list of predictions
		predictions.append(yhat)
		# add actual observation to history for the next loop
		history.append(test[i])
	# estimate prediction error
	error = measure_rmse(test, predictions)
	print(' > %.3f' % error)
	return error
```

```python
# repeat evaluation of a config
def repeat_evaluate(data, config, n_test, n_repeats=30):
	# fit and evaluate the model n times
	scores = [mlp_walk_forward_validation(data, n_test, config) for _ in range(n_repeats)]
	return scores
```

#### MLP model 
A simple grid search of model hyperparameters was performed and the configuration below was chosen. This may not be an optimal configuration, but is the best that was found.

- n_input: 24 (e.g. 24 months)
- n_nodes: 500
- n_epochs: 100
- n_batch: 100

```python
# data split
n_test = 12
# define config
config = [24, 500, 100, 100]
# grid search
data = series.values
scores = repeat_evaluate(data, config, n_test, n_repeats=10)
# summarize scores
summarize_scores('mlp', scores)
```
